```python
df_gpa = pd.read_csv("csv/gpa/gpa-dataset.csv")
df_gpa['Label'] = df_gpa["Subject"] + " " + df_gpa["Number"].astype(str)
```

```python
#https://github.com/illinois/reddit-uiuc-bot/blob/master/process_reddit_post.py
def get_recent_average_gpa(course):
    df = df_gpa[df_gpa["Label"] == course].groupby("Label").agg("sum").reset_index()
    if len(df) == 0: return None

    df["Count GPA"] = df["A+"] + df["A"] + df["A-"] + df["B+"] + df["B"] + df["B-"] + df["C+"] + df["C"] + df["C-"] + df["D+"] + df["D"] + df["D-"] + df["W"]
    df["Sum GPA"] = (4 * (df["A+"] + df["A"])) + (3.67 * df["A-"]) + \
    (3.33 * df["B+"]) + (3 * df["B"]) + (2.67 * df["B-"]) + \
    (2.33 * df["C+"]) + (2 * df["C"]) + (1.67 * df["C-"]) + \
    (1.33 * df["D+"]) + (1 * df["D"]) + (0.67 * df["D-"])

    df["Average GPA"] = df["Sum GPA"] / df["Count GPA"]
    return df["Average GPA"].values[0]
```

```python
round(get_recent_average_gpa("AAS 100"), 2)
```

```python
def map_gpas(row):
    recent_gpa = get_recent_average_gpa(row["label"])
    recent_gpa = "Not Available" if recent_gpa is None else str(round(recent_gpa, 2))
    row["gpa"] = recent_gpa 
    return row
```
