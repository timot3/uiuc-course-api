```python
def set_df_labels(df):
    df['Label'] = df['Subject'] + " " + \
        df['Number'].astype(str)
    return df
```

```python
df_fa22 = set_df_labels(pd.read_csv("./csv/2022-fa.csv"))
df_sp22 = set_df_labels(pd.read_csv("./csv/2022-sp.csv"))
df_sp21 = set_df_labels(pd.read_csv("./csv/2021-sp.csv"))
df_fa21 = set_df_labels(pd.read_csv("./csv/2021-fa.csv"))
df_sp20 = set_df_labels(pd.read_csv("./csv/2020-sp.csv"))
df_fa20 = set_df_labels(pd.read_csv("./csv/2020-fa.csv"))
df_sp19 = set_df_labels(pd.read_csv("./csv/2019-sp.csv"))
df_fa19 = set_df_labels(pd.read_csv("./csv/2019-fa.csv"))
df_sp18 = set_df_labels(pd.read_csv("./csv/2018-sp.csv"))
df_fa18 = set_df_labels(pd.read_csv("./csv/2018-fa.csv"))
```

```python
def preprocess_df(df) -> pd.DataFrame:
    df.columns = [x.lower().replace(' ', '_') for x in df.columns]
    return df
```

```python
df_all = preprocess_df(df_data)
df_all
```

```python
df_all["gpa"] = "None"
df_all["yearterm"] = df_all["term"] + " " + df_all["year"].astype(str)
```

```python
df_gpa = pd.read_csv("csv/gpa/gpa-dataset.csv")
df_gpa['Label'] = df_gpa["Subject"] + " " + df_gpa["Number"].astype(str)
```

```python
#https://github.com/illinois/reddit-uiuc-bot/blob/master/process_reddit_post.py
def get_recent_average_gpa(course):
    df = df_gpa[df_gpa["Label"] == course].groupby("Label").agg("sum").reset_index()
    if len(df) == 0: return None

    df["Count GPA"] = df["A+"] + df["A"] + df["A-"] + df["B+"] + df["B"] + df["B-"] + df["C+"] + df["C"] + df["C-"] + df["D+"] + df["D"] + df["D-"] + df["W"]
    df["Sum GPA"] = (4 * (df["A+"] + df["A"])) + (3.67 * df["A-"]) + \
    (3.33 * df["B+"]) + (3 * df["B"]) + (2.67 * df["B-"]) + \
    (2.33 * df["C+"]) + (2 * df["C"]) + (1.67 * df["C-"]) + \
    (1.33 * df["D+"]) + (1 * df["D"]) + (0.67 * df["D-"])

    df["Average GPA"] = df["Sum GPA"] / df["Count GPA"]
    return df["Average GPA"].values[0]
```

```python
def map_gpas(row):
    recent_gpa = get_recent_average_gpa(row["label"])
    recent_gpa = "Not Available" if recent_gpa is None else str(round(recent_gpa, 2))
    row["gpa"] = recent_gpa 
    return row
```

```python
df_n = df_all.progress_apply(map_gpas, axis=1)
```
